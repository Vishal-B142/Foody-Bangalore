```python
rest_type_mode = Za['rest_type'].mode()
cuisines_mode = Za['cuisines'].mode()
Za['cuisines'] = Za['cuisines'].fillna(cuisines_mode[0])
Za['rest_type'] = Za['rest_type'].fillna(rest_type_mode[0])
```

```python
Za.isnull().sum()
```

```python
Za = Za.drop(columns=['phone','location','dish_liked','menu_item'])
```

## Preprocessing Rate

```python
Za['rate'].unique()
```

```python
getting_rate = []
for i in Za['rate']:
    try:
        number = float(str(i[0:3]))
        getting_rate.append(number)
    except:
        getting_rate.append(np.nan)
        
Za[['rate']]=getting_rate
```

```python
Za.info()
```

# Preprocessing for price

```python
Za['approx_cost(for two people)']  
```

```python
getting_price = []
for s in Za['approx_cost(for two people)']:
    try:
        if type(s)==str:
            s = s.replace(',','')
        number = float(s)
        getting_price.append(number)
    except:
        getting_price.append(np.nan)

Za[['approx_cost(for two people)']]=getting_price     
    
```

```python
finallist = []
for i in range(len(Za)):
    temp = Za['cuisines'][i].split(',')
    u = [x.strip() for x in temp if x.strip()]
    for j in range(len(u)):
        finallist.append(u[j])
finalseries = pd.Series(finallist)
count = []
for cuisine in finalseries.unique():
    count.append(finallist.count(cuisine))
unique_cuisine_lst = list(finalseries.unique())
unique_cuisines_df = pd.DataFrame({'Unique_Cuisines': unique_cuisine_lst, 'Count': count})
unique_cuisines_df
```

```python
temp_count = max(unique_cuisines_df['Count'])
max_cuisine_count = unique_cuisines_df[unique_cuisines_df['Count'] == temp_count]
max_cuisine_count
```

```python
plt.figure(figsize=(80, 40))
b = sns.barplot(x = 'Unique_Cuisines', y = 'Count', data = unique_cuisines_df)
plt.title('Count of Each Cusines')
b.set_xticklabels(b.get_xticklabels(), fontsize = 20, rotation = 90)
plt.show()
```
